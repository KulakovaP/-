# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Дано:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
numbers = input("Введите список чисел через пробел ").split()
sum_numbers = 0
max_number = numbers[0]
for number in numbers:
    sum_numbers += int(number)
    if number > max_number:
        max_number = number
average = sum_numbers / len(numbers)
print("Сумма чисел", sum_numbers)
print("Среднее значение", average)
print("Максимальное значение", max_number)


````

# **Задача 3**

Простой чат-бот
Создайте простого чат-бота, который отвечает на основные запросы пользователей. Программа должна использовать условные выражения для определения намерений пользователя и последующего соответствующего ответа.

Чат бот может:

- помогать складывать числа
- подсказывать фильм, музыку и тд (из заранее созданного списка)
- создавать список дел (а также выводить его, обновлять, очищать)

Интерфейс чат-бота на ваше усмотрение, это могут быть как заранее определенные команды, так и к примеру вызов команды по номеру

*Реализовать без использования функций*

Пример:

Ввод:

```
Олег, какую мызыку посоветуешь?
```

Вывод:

```
Я бы посоветовал вам: Led Zeppelin Stairway to Heaven
```

````python
import random
movies = ["Побег из Шоушенка", "Властелин колец", "Зелёная миля", "Звёздные войны", "Назад в будущее", "Унесённые призраками", "Король лев", "Человек-паук", "1+1", "Чужой", "Ходячий замок", "Гарри Поттер", "Челюсти", "Хатико", "Поймай меня, если сможешь", "Титаник"]
music = ["Taylor Swift", "Lady Gaga", "Justin Timberlake", "Jennifer Lopez", "Bea Miller", "BTS", "Billie Eilish", "Rihanna"]

todo_list = []


while True:
    user_input = input('Что Вы хотите сделать? 1.Выход 2.сложить числа 3.Подсказать фильм 4.Подсказать музыку 5.Список дел 6.создать 7.отчистить')

    if user_input == '1':
        break

    elif user_input == '2':
        num1 = int(input('Введите первое число'))
        num2 = int(input('Введите второе число'))
        print('Результат:', num1 + num2)

    elif user_input == '3':
        print('Можете посмотреть этот фильм: ', random.choice(movies))

    elif user_input == '4':
        print('Можете послушать этого исполнителя: ', random.choice(music))

    elif user_input == '5':
        if not todo_list:
            print('Список дел пуст')
        else:
            print('Ваш список дел:')
            for item in todo_list:
                print('-', item)

    elif user_input == '6':
        item = input('Введите задачу ')
        todo_list.append(item)
        print('Задача', item, 'добавлена в список дел')

    elif user_input == '7':
        todo_list.clear()
        print('Список дел очищен')

    else:
        print('Попробуйте другой вариант')
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random
def play(a, b):
  if a == b:
    return 'ничья'
  elif (a == 'камень' and b == 'ножницы') or (a == 'ножницы' and b == 'бумага'):
    return 'человек'
  else:
    return 'компьютер'


player1 = 0    #человек
player2 = 0     #компьютер
while (player1 < 3) and (player2 < 3):
  human = input('Введите выбор  ')
  vab = ['камень', 'ножницы', 'бумага']
  computer = random.choice(vab)
  print(computer)
  hod = play(human, computer)
  if hod == 'ничья':
    print('У вас ничья')
    print(player1, '/', player2)
  elif hod == 'человек':
    print('Вы выйграли')
    player1 += 1
    print(player1, '/', player2)
  else:
    player2 += 1
    print('Компьютер выйграл')
    print(player1, '/', player2)

if player1 == 3:
  print('Вы победили!')
else:
  print('Компьютер победил(')


````

# **Задача 6.**

Создайте игру "Палач" или "Виселица", в которой пользователь должен угадать слово, предлагая буквы. Программа должна выбрать случайное слово, а затем разрешить пользователю угадывать буквы. После каждого угадывания программа должна сообщать пользователю, есть ли в слове буква или нет.

**Этап 1**
Создать саму игру

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
import random

words = ["python", "программирование", "игра", "палач", "виселица"]
word_to_guess = random.choice(words)
guessed_letters = []
attempts = 6  # количество попыток
score = 0

print('Угадайте слово, предлагая буквы. У Вас есть', attempts, 'попыток')

while attempts > 0:
    # Отображение текущего состояния слова
    current_state = ''.join([letter if letter in guessed_letters else '_' for letter in word_to_guess])
    print("Слово:", current_state)

    # Проверка, угадано ли слово
    if current_state == word_to_guess:
        print("Поздравляем! Вы угадали слово:", word_to_guess)
        score += 1
        break

    # Ввод буквы или слова
    guess = input("Введите букву или слово (или 'exit' для выхода): ").lower()

    if guess == "exit":
        print("Вы вышли из игры.")
        break

    # Проверка, является ли ввод буквой
    if len(guess) == 1 and guess.isalpha():
        if guess in guessed_letters:
            print("Вы уже угадали эту букву.")
        else:
            guessed_letters.append(guess)
            if guess in word_to_guess:
                print("Правильно! Буква", guess, "есть в слове.")
            else:
                attempts -= 1
                print("Неправильно! Буквы", guess, "нет в слове. Осталось попыток:", attempts)
    elif len(guess) == len(word_to_guess) and guess.isalpha():
        if guess == word_to_guess:
            print("Поздравляем! Вы угадали слово:", word_to_guess)
            score += 1
            break
        else:
            attempts -= 1
            print("Неправильно! Осталось попыток:", attempts)
    else:
        print("Некорректный ввод. Пожалуйста, введите одну букву или слово.")

if attempts == 0:
    print("Вы проиграли! Загаданное слово было:", word_to_guess)

print("Ваш счет:", score)
````
